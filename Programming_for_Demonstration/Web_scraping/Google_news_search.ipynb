```python
from pygooglenews import GoogleNews
import pandas as pd
search_result = []
search_exclude = ['人民網','网','新浪','中国','澳門','华','搜狐','闻','hk','香港','cn','大陸']
counter = 0
gn = GoogleNews(country = 'TW', lang = 'zh-TW')
search = gn.search('intitle:\"科技部\"', from_ = '2021-08-16', to_ = '2021-08-31')
search['entries']
#loop the html structure.
for news in search['entries']:
    flag = True
    for word in search_exclude:
        #checking the unwanted keywords.
        if word in news['title'] or word in news['link']:
            flag = False
            break
    #put into the lists.
    if flag == True:
        
        print(news['title'])
        print(news['link'])
        print(news['source']['title'])
        print(news['published'])
        print('----------------------------')
        search_result.append([])
        search_result[counter].append(news['title'])
        search_result[counter].append(news['link'])
        search_result[counter].append(news['source']['title'])
        search_result[counter].append(news['published'])
        
        counter += 1
        
df = pd.DataFrame(search_result, columns = ['標題', '連結', '新聞媒體', '日期'])
df['日期'] = pd.to_datetime(df.日期)
df = df.sort_values(by = '日期', ascending = True)
df['日期'] = df['日期'].astype(str)
#count the numbers of news.
print(len(search['entries']))
print(counter)
print(search_result)
print('df的資料結構是{}'.format(df.shape))
print(df)
```
